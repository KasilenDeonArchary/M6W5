```python
# M6W5.


# Import required libraries.
import torch
from torch import nn, optim
import torch.nn.functional as F
from torchvision import datasets, transforms
import matplotlib.pyplot as plt
import numpy as np




# Check for GPU.
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")


# Normalize the data.
transform = transforms.Compose([transforms.ToTensor(), transforms.Normalize((0.5,), (0.5,))])


# Download and load training data.
trainset = datasets.FashionMNIST('~/.pytorch/F_MNIST_data/', download=True, train=True, transform=transform)
trainloader = torch.utils.data.DataLoader(trainset, batch_size=64, shuffle=True)


# Download and load test data.
testset = datasets.FashionMNIST('~/.pytorch/F_MNIST_data/', download=True, train=False, transform=transform)
testloader = torch.utils.data.DataLoader(testset, batch_size=64, shuffle=True)


class AutoEncoder(nn.Module):
    def __init__(self, architecture):
        super(AutoEncoder, self).__init__()

        if architecture == 1:
# Encoder.
            self.enc1 = nn.Linear(in_features=784, out_features=256)
            self.enc2 = nn.Linear(in_features=256, out_features=64)
# Decoder.
            self.dec1 = nn.Linear(in_features=64, out_features=256)
            self.dec2 = nn.Linear(in_features=256, out_features=784)
            
        elif architecture == 2:
# Encoder.
            self.enc1 = nn.Linear(in_features=784, out_features=256)
            self.enc2 = nn.Linear(in_features=256, out_features=128)
            self.enc3 = nn.Linear(in_features=128, out_features=64)
# Decoder.
            self.dec1 = nn.Linear(in_features=64, out_features=128)
            self.dec2 = nn.Linear(in_features=128, out_features=256)
            self.dec3 = nn.Linear(in_features=256, out_features=784)
            
        elif architecture == 3:
# Encoder.
            self.enc1 = nn.Linear(in_features=784, out_features=256)
            self.enc2 = nn.Linear(in_features=256, out_features=64)
# Decoder.
            self.dec1 = nn.Linear(in_features=64, out_features=256)
            self.dec2 = nn.Linear(in_features=256, out_features=784)
            
        self.architecture = architecture

    def forward(self, x):
        x = x.view(-1, 784)
        
        if self.architecture == 1:
            # Encoder
            x = F.relu(self.enc1(x))
            x = F.relu(self.enc2(x))
# Decoder.
            x = F.relu(self.dec1(x))
            x = torch.tanh(self.dec2(x))
            
        elif self.architecture == 2:
# Encoder.
            x = F.relu(self.enc1(x))
            x = F.relu(self.enc2(x))
            x = F.relu(self.enc3(x))
# Decoder.
            x = torch.sigmoid(self.dec1(x))
            x = torch.sigmoid(self.dec2(x))
            x = torch.tanh(self.dec3(x))
            
        elif self.architecture == 3:
            # Encoder
            x = torch.sigmoid(self.enc1(x))
            x = torch.sigmoid(self.enc2(x))
            # Decoder
            x = torch.sigmoid(self.dec1(x))
            x = torch.tanh(self.dec2(x))
            
        return x

    
def train_model(model, trainloader, epochs, optimizer, criterion):
    losses = []
    for epoch in range(epochs):
        running_loss = 0
        for images, _ in trainloader:
            images = images.to(device)
            optimizer.zero_grad()
            outputs = model(images)
            loss = criterion(outputs, images.view(-1, 784))
            loss.backward()
            optimizer.step()
            running_loss += loss.item()

        losses.append(running_loss/len(trainloader))
        print(f"Epoch {epoch+1} - Training loss: {running_loss/len(trainloader)}")

        
# Plot of loss vs time.
    plt.plot(losses)
    plt.show()

    
def visualize_reconstruction(model, testloader):
    for images, _ in testloader:
        images = images.to(device)
        output = model(images)
        output = output.view(output.size(0), 1, 28, 28)
        output = output.cpu().detach().numpy()
        input_images = images.view(images.size(0), 1, 28, 28)
        input_images = input_images.cpu().numpy()
# For 1 batch.        
        break  

    fig, axes = plt.subplots(nrows=2, ncols=10, sharex=True, sharey=True, figsize=(25,4))
    for images, row in zip([input_images, output], axes):
        for img, ax in zip(images, row):
            ax.imshow(np.squeeze(img), cmap='gray')
            ax.get_xaxis().set_visible(False)
            ax.get_yaxis().set_visible(False)
    plt.show()

    
for architecture in range(1, 4):
    print(f"Training AE architecture {architecture}")
    model = AutoEncoder(architecture).to(device)
    optimizer = optim.Adam(model.parameters(), lr=0.001)
    criterion = nn.MSELoss()

    train_model(model, trainloader, epochs=20, optimizer=optimizer, criterion=criterion)
    visualize_reconstruction(model, testloader)
    
    
# Select architecture.
architecture = 1


# Model, optimizer and loss function.
model = AutoEncoder(architecture).to(device)
optimizer = optim.Adam(model.parameters(), lr=0.001)
criterion = nn.MSELoss()


# Train model.
train_model(model, trainloader, epochs=20, optimizer=optimizer, criterion=criterion)


# Visualize the reconstructions on test data.
visualize_reconstruction(model, testloader)




# I selected architecture 1 as the final model because of it's simplicity and performance. 
# Both architecture 1 and architecture 3 have the same number of layers and neurons, but architecture 1 
# uses ReLU activation functions while architecture 3 uses sigmoid functions. Since the ReLU activation 
# function is efficient and does not suffer from vanishing gradient problem, architecture 1 performs better.


# Autoencoders can be used to generate data that closely resembles the input data. This is what 
# "creating new clothes" means. I used the autoencoder to generate images that should closely
# resemble the input images.


# Accuracy is not usually a metric used to evaluate the performance of autoencoders. 
# Autoencoders are unsupervised learning models, so I do not have target labels to compare the model's 
# predictions to. I instead looked at how well the autoencoder can recreate its input data, which I could
# measure using a reconstruction loss such as Mean Squared Error (MSE).

```
